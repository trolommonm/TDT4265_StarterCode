```python
import sys, os
sys.path.append(os.path.dirname(os.getcwd())) # Include ../SSD in path
import tops
from tops.config import instantiate
import pathlib
from ssd.utils import load_config
from plot_utils import plot
```

```python
task24_logs = tops.logger.read_logs(pathlib.Path("./outputs/task24_config_5/"))
task41_2_logs = tops.logger.read_logs(pathlib.Path("./outputs/task41_config_2/"))
task41_3_logs = tops.logger.read_logs(pathlib.Path("./outputs/task41_config_3/"))

list_of_logs = [task24_logs, task41_2_logs, task41_3_logs]
labels = ["Approach 1", "Approach 2", "Approach 3"]
```

```python
# plot mAP@0.5:0.95
fig, ax = plot(list_of_logs, labels, "metrics/mAP")
fig.legend(loc="center right")
```

```python
print("Highest mAP@0.5:0.95 reach for Approach 1: ", max([entry["metrics/mAP"] for entry in task24_logs if "metrics/mAP" in entry]))
print("Highest mAP@0.5:0.95 reach for Approach 2: ", max([entry["metrics/mAP"] for entry in task41_2_logs if "metrics/mAP" in entry]))
print("Highest mAP@0.5:0.95 reach for Approach 3: ", max([entry["metrics/mAP"] for entry in task41_3_logs if "metrics/mAP" in entry]))
```

```python
# plot regression loss
fig, ax = plot(list_of_logs, labels, "loss/regression_loss")
fig.legend()
```

```python
# plot classification loss
fig, ax = plot(list_of_logs, labels, "loss/classification_loss")
fig.legend()
```

```python
# plot total loss
fig, ax = plot(list_of_logs, labels, "loss/total_loss")
fig.legend()
```
